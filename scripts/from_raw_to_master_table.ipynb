Used for this is spacers.filtered.extra file. This one is created from the spacercount file. Which is created by the scripts/extract_spacercount.zsh, which extracts from the collect folder, created by the collect.v2.bash script. But also combine the virsorter data, and the 'all_spacers' file, which is all the blast lines of the selfhits of spacers. 

```python
df = pd.read_csv('/home/hielke/bep/combined_data/all.filtered.spacers',
                 names=['genome', 'spacer', 'c_1', 'c_2', 'contig_hit', 'length', 'ident', 'total_spacers', 'prophage', 'has_phages'], 
                 sep='\t', 
                 dtype={'genome': str, 'spacer': str, 'length': int, 'ident': float, 'c_1': int, 'c_2': int, 'genome_hit': str, 'total_spacers': int, 'prophage': str})
df.head()
```

```python
dfg = pd.read_csv('/home/hielke/bep/data/selfhits/PATRIC/PATRIC_genomes',
                 names=["genome", "species", "genus"],
                 sep="\t",
                 dtype={"genome": str, "species": str, "genus": str},)
dfg.head()
```

```python
# del df['c_1']
# del df['c_2']
```

```python
df.set_index('genome', inplace=True)
dfg.set_index("genome", inplace=True)
df.head()
```

```python
dfd = pd.read_csv('/home/hielke/bep/data/disco/discodata/MasterMaster', 
                  dtype={'Total non V-U Prots': int, 'Total Cas Prots': int}, index_col=0)
dfd.head()
```

```python
dfd.rename(columns={'organism': 'genome'}, inplace=True)
dfd['genome'] = dfd['genome'].map(lambda x : ".".join(x.split('-')))
dfd.set_index('genome', inplace=True)
dfd.head()
```

```python
combined = df.combine_first(dfd)
combined = combined.combine_first(dfg)
combined.head()
```

```python
combined.to_csv('fp_combined_with_extra.tsv', sep='\t')
```
